## 문제2. 텍스트 데이터셋에서 단어를 추출합니다. 

```python
import collections #이거 안배웠는데 진짜 저자 때릴까.

corpus = " ".join(df['Description'].tolist()).split(" ")
count = collections.Counter(corpus)
print(count)
```

## 문제3. 추출한 단어를 이용하여 워드 클라우드를 출력합니다. 

```python
import random
import pytagcloud
import webbrowser
```

```python
#출현 빈도가 가장 높은 40개의 단어를 선정합니다.
ranked_tags = count.most_common(40)

#pytagcloud로 출력할 40개의 단어를 입혁합니다. 단어 출력의 최대 크기는 200으로 제한합니다.
taglist = pytagcloud.make_tags(ranked_tags, maxsize = 200)
```

```python
#pytagcloud 이미지를 생성합니다. 폰트는 나눔 고딕을 사용합니다.

pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(900,600),
fontname = 'NanumGothic', rectangular = False)
```
